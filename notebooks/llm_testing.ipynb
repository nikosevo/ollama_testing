```python
import requests
import json
```

```python
promt = 'thats cool but i need it in an html format plus some kind of tailwind css  '
# Make the POST request
res = requests.post(
    "http://localhost:11434/api/generate",
    json={"model": "phi3", "prompt": promt},
    stream=True  # enable streaming
)

full_text = ""

# Read the response line by line
for line in res.iter_lines():
    if line:
        chunk = json.loads(line)
        full_text += chunk.get("response", "")

print("AI:", full_text)
```
